```python
import pandas as pd

df = pd.read_csv("data/online_retail.csv", encoding="latin1")
df.head()
```

```python
df = df.dropna(subset=["CustomerID"])
```

```python
df = df.dropna(subset=["Description"])
```

```python
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]
```

```python
df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
df = df.drop_duplicates()
```

```python
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
```

```python
df["CustomerID"] = df["CustomerID"].astype(int)
```

```python
df.shape
df.isnull().sum()
df.duplicated().sum()
df.head()
```

```python
# Latest date in the dataset + 1 day (for recency calculation)
snapshot_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)

rfm = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (snapshot_date - x.max()).days,  # Recency
    "InvoiceNo": "nunique",  # Frequency = number of unique transactions
    "TotalPrice": "sum"      # Monetary = total spending
})

rfm.columns = ["Recency", "Frequency", "Monetary"]
rfm.head()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(rfm["Recency"], bins=50, kde=True)
plt.title("Recency Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(rfm["Frequency"], bins=50, kde=True)
plt.title("Frequency Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(rfm["Monetary"], bins=50, kde=True)
plt.title("Monetary Distribution")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

```python
rfm.shape
```

```python
from sklearn.cluster import KMeans

wcss = []

K_range = range(2, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10,4))
plt.plot(K_range, wcss, marker="o")
plt.title("Elbow Curve for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Inertia)")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(rfm_scaled)
    score = silhouette_score(rfm_scaled, labels)
    print(f"K={k}  Silhouette Score={score:.4f}")
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)

rfm.head()
```

```python
cluster_profile = rfm.groupby("Cluster")[["Recency","Frequency","Monetary"]].mean()
cluster_profile
```

```python
segment_map = {
    0: "High-Value",
    1: "Regular",
    2: "Occasional",
    3: "At-Risk"
}

rfm["Segment"] = rfm["Cluster"].map(segment_map)
rfm[["Recency","Frequency","Monetary","Cluster","Segment"]].head()
```

```python
cluster_counts = rfm["Segment"].value_counts()

plt.figure(figsize=(8,4))
sns.barplot(x=cluster_counts.index, y=cluster_counts.values)
plt.title("Customer Segments Count")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
cluster_profile
```

```python
final_segment_map = {
    2: "High-Value",
    3: "Regular",
    0: "Occasional",
    1: "At-Risk"
}

rfm["Segment"] = rfm["Cluster"].map(final_segment_map)
rfm[["Recency","Frequency","Monetary","Cluster","Segment"]].head()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x=rfm["Segment"], order=["High-Value","Regular","Occasional","At-Risk"])
plt.title("Customer Segmentation    ")
plt.xlabel("Segment")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df["Description"] = df["Description"].astype(str).str.strip()
```

```python
pivot = df.pivot_table(
    index="Description",
    columns="CustomerID",
    values="Quantity",
    aggfunc="sum",
    fill_value=0
)

pivot.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

product_similarity = cosine_similarity(pivot)
similarity_df = pd.DataFrame(product_similarity, index=pivot.index, columns=pivot.index)

similarity_df.iloc[:5, :5]
```

```python
def recommend_products(product_name, similarity_df, top_n=5):
    if product_name is None or product_name.strip() == "":
        return []

    # normalize input
    product_name_clean = product_name.strip().upper()

    # build uppercase mapping for matching
    product_map = {p.upper(): p for p in similarity_df.index}

    if product_name_clean not in product_map:
        return None  # not found

    actual_name = product_map[product_name_clean]

    # similarity scores
    scores = similarity_df[actual_name].sort_values(ascending=False)

    # remove the same product from recommendations
    recommendations = scores.iloc[1:top_n+1].index.tolist()
    return recommendations
```

```python
recommend_products("WHITE METAL LANTERN", similarity_df)
```
